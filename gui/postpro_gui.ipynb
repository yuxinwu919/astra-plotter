# ASTRA PostPro GUI — 交互式后处理界面

基于 ipywidgets 的交互式 ASTRA 数据可视化工具，复现 ASTRA 自带 postpro 程序的核心功能。

**功能列表：**
- 📁 文件浏览器选择数据目录
- 📊 发射度 / 束斑 / 能量演化曲线
- 🎨 相空间密度图（含投影）
- 🔬 切片分析 & 亮度计算
- 💾 图表导出为 PDF/PNG
- 📦 一键数据备份

```python
# ============================================================
# 初始化
# ============================================================
import os
import sys

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

from gui import PostProGUI, build_jupyter_gui, build_phase_space_gui
from astra_plotter.core.cosmetics import set_publication_style

# 设置绘图风格
set_publication_style(font_size=14, use_tex=False)
%matplotlib inline

print("✓ GUI 模块加载成功")
```

```python
# ============================================================
# 选择数据目录并加载（自动检测项目名称）
# ============================================================

# 自动检测默认数据目录
DEFAULT_DATA_DIR = os.path.join(PROJECT_ROOT, 'simulation_output')
DATA_DIR = DEFAULT_DATA_DIR  #NOTE 可手动修改此路径

print(f"数据目录: {DATA_DIR}")

# 自动检测 ASTRA 项目根名称
from astra_plotter.core.loader import auto_detect_rootname
detected_rootname = auto_detect_rootname(DATA_DIR)

if detected_rootname:
    print(f"🔍 自动检测项目名称: '{detected_rootname}'")
else:
    detected_rootname = 'astra'
    print("⚠ 未检测到 ASTRA 文件，使用默认名称 'astra'")

# 创建 GUI 实例（自动使用检测到的项目名称）
gui = PostProGUI(data_dir=DATA_DIR, rootname=detected_rootname, run='001')

# 加载所有数据
gui.load_data()

# 显示统计信息
gui.show_statistics()

# 运行切片分析
gui.run_slice_analysis(n_slices=30)
```

## 演化曲线

发射度、束斑尺寸、能量、切片分析等。横坐标为 $z$ 或 wavenumber，不依赖束团位置。

```python
# 构建并显示交互式 GUI
gui_widget = build_jupyter_gui(gui)
display(gui_widget)
print("\n💡 选择绘图类型后点击「生成图表」按钮")
```

## 相空间

横向/纵向相空间密度图，需要在「束团位置」下拉菜单中选择对应位置的相空间文件。

```python
# 构建相空间交互面板（含束团位置选择器）
phase_gui_widget = build_phase_space_gui(gui)
display(phase_gui_widget)
print("\n💡 先选择束团位置，再选择绘图类型并点击「生成图表」")
#TODO 统计信息里发射度的单位需要确认一下，pi·mm·mrad 还是 mm·mrad
```
